```python
import cv2
import os, glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Extraction Only White Colors From the Image
```

```python
def extract_white(image):
    #color Mask
    low=np.uint8([200,200,200])
    high=np.uint8([255,255,255])
    white_mask=cv2.inRange(image,low,high)
    output=cv2.bitwise_and(image,image , mask=white_mask)
    
    #show Image
    #cv2.imshow("images",np.hstack([image,output]))
    plt.imshow(output)
    
```

```python
img=cv2.imread('i_m.jpg',1)
```

```python
plt.imshow(img)
```

```python
extract_white(img)
```
